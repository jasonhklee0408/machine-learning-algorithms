```python
#8
heart = pd.read_csv(r"C:\Users\jason\Desktop\cse 151a\hw4\heart.csv")
heart
```

```python
y = heart['target']
X = heart.drop(['target'],axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size= 200, random_state=42)
```

```python
lr = LogisticRegression(max_iter=10000).fit(X_train, y_train)
predictions = lr.predict(X_test)
```

```python
#8a
coefs1 = np.abs(lr.coef_[0])
np.array(X_train.columns)[np.argsort(coefs1)[-3:]]
```

```python
#8b
np.sum(y_test.values != predictions)/len(predictions)
```

```python
#8c
kf = KFold(n_splits = 5)
error = []
for trainIndex, testIndex in kf.split(X_train):
    xTrain = X_train.iloc[trainIndex]
    xTest = X_train.iloc[testIndex]
    lr = LogisticRegression(max_iter=10000).fit(xTrain, y_train.values[trainIndex])
    predictions = lr.predict(xTest)
    error.append(np.sum(y_train.values[testIndex] != predictions)/xTest.shape[0])

np.mean(error)
```
